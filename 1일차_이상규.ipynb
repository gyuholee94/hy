# 가위바위보 게임
# 사용자값은 입력받고 컴퓨터값은 무작위로 생성한뒤
# 승부를 판단하고 결과를 보여준다. 

```python
import random
import time

winner = '컴퓨터'
time_interval=0.1

# 게임 인트로 -------------------------------------------------------
print ('가위~', end=' ')
time.sleep(time_interval*5)
print ('바위~', end=' ')
time.sleep(time_interval*5)
print ('보~!!!!! ')
time.sleep(time_interval*5)

print ('가위~', end=' ')
time.sleep(time_interval*2)
print ('바위~', end=' ')
time.sleep(time_interval*2)
print ('보~!!!!! ')
time.sleep(time_interval*2)

while winner!='당신':

# 사용자 결정-------------------------------------------------------
    yourHand = input('가위~ 바위~ 보 ! 중에 하나를 골라주세요~   ')
    print ()
    
# 컴퓨터 결정-------------------------------------------------------
    comRandom = random.randint(1,3)
    if comRandom == 1:
        comHand = '가위'
    elif comRandom == 2:
        comHand = '바위'
    else:
        comHand = '보'
        
# 승부판단-------------------------------------------------------
    if yourHand == '가위':
        yourNum = 1
        if yourNum == comRandom:
            winner = 0
        elif comRandom-yourNum>1:
            winner = '당신'
        else: winner = '컴퓨터'

    elif yourHand == '바위':
        yourNum = 2
        if yourNum == comRandom:
            winner = 0
        elif yourNum>comRandom:
            winner = '당신'
        else: winner = '컴퓨터'

    else :
        yourNum = 3
        if yourNum == comRandom:
            winner = 0
        elif yourNum-comRandom>1:
            winner = '컴퓨터'
        else: winner = '당신'
            
# 출력-------------------------------------------------------
    if winner == '당신':
        print ('당신 이김! 컴퓨터 짐! (You:' + yourHand + '~, Com:' + comHand + '~)')
    elif winner == '컴퓨터':
        print ('당신 짐! 컴퓨터 이김! 또 해! (You:' + yourHand + '~, Com:' + comHand + '~)')
    else :
        print ('무승부! 또 해! (You:' + yourHand + '~, Com:' + comHand + '~)')

    print()
    print()
    time.sleep(1.5)
```
